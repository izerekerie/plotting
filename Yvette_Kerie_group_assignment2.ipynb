```python
plt.figure(figsize=(8, 5))
plt.hist(df['co2'], bins=25, color='green')
plt.title('Distribution of carbon Dioxide')
plt.show()
```

```python
#This histogram is slightly skewed to the left and most values cluster around 27 to 35 MPG, with a sharp drop-off below 20 and above 40 MPG.
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['displ'], bins=25, color='green')
plt.title('Distribution of displ')
plt.show()
```

```python
# The distribution is highly right-skewed and most values range between 1.5 and 4.0 liters, with distinct peaks at 2.0 L, 3.0 L, and 3.5 L.
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['highway'], bins=25, color='green')
plt.title('Distribution of highway')
plt.show()
```

```python
# The histogram is slightly right-skewed and Most values fall between 25-40 with a peak around 28-30.
```
